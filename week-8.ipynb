# Logistic Regression for binary classification

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# -----------------------------
# 1. Dataset   You can read csv file also
# -----------------------------
df = pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Play Tennis.csv")
df.head()
```

```python
# -----------------------------
# 2. Preprocessing
# -----------------------------
X = df.drop(columns=['Day',"Play_Tennis"])   # features
y = df['Play_Tennis']                        # target
```

```python
# One-hot encode categorical features
X=pd.get_dummies(X)
X
```

```python
# Encode target variable (Yes=1, No=0)
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
# Split into Train/Test
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# -----------------------------
# 3. Logistic Regression Model
# -----------------------------
lgr=LogisticRegression()
model=lgr.fit(X_train,y_train)
```

```python
# Predictions
y_pred=model.predict(X_test)
```

```python
# -----------------------------
# 4. Evaluation
# ----------------------------- 
print("The Accuracy is",metrics.accuracy_score(y_test,y_pred))
print("The Confusion metrix is \n",metrics.confusion_matrix(y_test,y_pred))
```

```python
print("The classification report is \n",metrics.classification_report(y_test,y_pred))
```

# Logistic Regression for multi class classification

```python
import pandas as pd
df = pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/dataset_Fish.csv")
df.head()
```

```python
x=df.drop("Species",axis=1)
y=df["Species"]
```

```python
x.head()
```

```python
y.head()
```

# Scaling the input features using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
ml=MinMaxScaler()
x=ml.fit_transform(x)
x[0:5]
```

# Label Encoding the target variable using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
y
```

# Splitting into train and test datasets using train_test_split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Building and training

```python
lgr=LogisticRegression()
# training the model
model=lgr.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

# Confusion Matrix ( You can also plot simple confusion matrix)

```python
print("Accuracy",metrics.accuracy_score(y_test,y_pred))
print("Confusion metrix \n",metrics.confusion_matrix(y_test,y_pred))
print("classification report \n",metrics.classification_report(y_test,y_pred))
```
