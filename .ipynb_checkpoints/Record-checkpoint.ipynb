# 3) Consider titanic datset and plot diffrent graphs to visualize Trends , Proposition ,Categorical data ,Unsertinetiese and Distrubutions

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading CSV File
df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/titanic.csv")
```

```python
# 1. TREND (Survival over Age)
df.groupby("age")["survived"].mean().plot()
plt.title("Survival Trend over Age")
plt.xlabel("Age")
plt.ylabel("Survival Rate")
plt.show()

```

```python
# 2. PROPORTION (Pie Chart of Class Distribution)
df["class"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Passenger Proportion by Class")
plt.ylabel("")
plt.show()
```

```python
# 3. CATEGORICAL DATA (Bar Chart - Survival by Sex)
class_counts = df["class"].value_counts()
plt.bar(class_counts.index, class_counts.values, color=["skyblue","orange","green"])
plt.title("Categorical Data: Passenger Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# 4. UNCERTAINTIES (Error Bars - Survival by Class)
df.boxplot(column="age", by="class")
plt.title("Age Distribution by Class")
plt.suptitle("")  # remove automatic subtitle
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()
```

```python
# 5. DISTRIBUTIONS (Histogram of Age)
plt.hist(df["age"].dropna(), bins=30, color="purple")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# a) Consider titanic dataset and handle the missing values using diffrent methods.

```python
#Reading CSV File
titanic=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/titanic.csv")
```

```python
# Check missing values
print(titanic.isnull().sum())
```

### 1. Remove rows with missing values

```python
titanic_drop = titanic.dropna()
print("Shape after dropping NA:", titanic_drop.shape)
```

### 2. Fill missing values with a constant

```python
titanic_fill_constant = titanic.fillna({"embarked": "Unknown", "deck": "U"})
```

### 3. Fill with Mean / Median (for numerical columns like Age)

```python
# Fill Age with mean
titanic_mean = titanic.copy()
titanic_mean["age"] = titanic_mean["age"].fillna(titanic_mean["age"].mean())
```

```python
# Fill Age with median
titanic_median = titanic.copy()
titanic_median["age"] = titanic_median["age"].fillna(titanic_median["age"].median())
```

### 4. Fill with Mode (for categorical columns)

```python
titanic_mode = titanic.copy()
titanic_mode["embarked"] = titanic_mode["embarked"].fillna(titanic_mode["embarked"].mode()[0])
```

### 5. Forward Fill / Backward Fill

```python
# Forward fill
titanic_ffill = titanic.fillna(method="ffill")

# Backward fill
titanic_bfill = titanic.fillna(method="bfill")
```

### Detect Outliers (IQR Method)

```python
Q1 = titanic["fare"].quantile(0.25)
Q3 = titanic["fare"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print("Lower Bound:", lower_bound, " Upper Bound:", upper_bound)

outliers = titanic[(titanic["fare"] < lower_bound) | (titanic["fare"] > upper_bound)]
print("Number of outliers in Fare:", len(outliers))
```

# Handle Outliers (Different Methods)

## 1. Remove Outliers

```python
titanic_no_outliers = titanic[(titanic["fare"] >= lower_bound) & (titanic["fare"] <= upper_bound)]
```

## Replace Outliers using Upper Cap and Lower Cap

```python
import numpy as np
fare_arr = df["fare"]
lower_cap = np.percentile(fare_arr,1)
upper_cap = np.percentile(fare_arr,99)
```
