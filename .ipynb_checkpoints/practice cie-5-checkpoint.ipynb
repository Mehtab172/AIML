```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# plot the 

plt.matshow(X_train[0])
plt.show()
```

```python
#Convert pixel values from the range 0‚Äì255 to 0‚Äì1 for better model performance.‚Äù

X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
from tensorflow import keras

model = keras.Sequential([
    keras.Input(shape=(784,)),          # üëà Input layer
    keras.layers.Dense(10, activation='softmax')  # üëà use softmax, not sigmoid
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# Evaluate the model ab
model.evaluate(X_test_flattened, y_test)
```

```python
# Predict the model

y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
# Show the predict plot 
plt.matshow(X_test[11])
```

```python
np.argmax(y_predicted[11])
```

```python
from sklearn import metrics
```

```python
y_predicted_classes = np.argmax(y_predicted, axis=1)
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_classes)
cm
```

```python
print("Classification Report:\n",metrics. classification_report(y_test, y_predicted_classes))
```

```python
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
