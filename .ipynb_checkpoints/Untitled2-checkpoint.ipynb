```python
import pandas as pd   
import numpy as np   
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Play Tennis.csv")
df.head()
```

```python
x=df.drop(["Day","Play_Tennis"],axis=1)
y=df["Play_Tennis"]
```

```python
lb=LabelEncoder()
y=lb.fit_transform(y)
```

```python
y
```

```python
x=pd.get_dummies(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lgs=LogisticRegression()
model=lgs.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("cofusion metrix fro this \n",metrics.confusion_matrix(y_test,y_pred))
print("accuracy is \n",metrics.accuracy_score(y_test,y_pred))
print("classification re[rt is \n",metrics.classification_report(y_test,y_pred))
```
