# 4) Handling Missing Values and Outliares

```python
df.isnull().sum()
```

```python
mean =df['Inches'].mean()
df['Inches'] =df['Inches'].fillna(mean)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Inches'].plot(kind='box')
plt.show()
```

```python
# finding the 1st quartile
q1 = df["Inches"].quantile(0.25)

# finding the 3rd quartile
q3 = df['Inches'].quantile(0.75)

# finding the iqr region
iqr = q3-q1

# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
```

```python
m = df['Inches'].mean()
print('mean:',m)
for i in df['Inches']:
    if i<lower_bound or i>upper_bound :
        df['Inches'] = df['Inches'].replace(i,m)
```

```python
df['Inches'].plot(kind='box')
plt.show()
```

6) Data Splitting and Training

```python
X = df.drop('Price', axis=1)   
y = df['Price']                
```

```python
# Convert categorical columns into numeric (One-Hot Encoding)
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=50)
```

```python
X_train
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python


y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print("RMSE",rmse)
```

```python
print("R2_score",r2)
```
