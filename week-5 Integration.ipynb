# Given the event dataset which contains data,event,cost columns, perform the following operations

```python
import pandas as pd 
data = {
    'Date':['10/2/2025','11/2/205','15/2/2025'],
    'Event':['music','poetry','comedy'],
    'Cost':[10000,5000,15000]}
df=pd.DataFrame(data)
df
```

# a) Add two new columns Discount and Discounted Price

```python
df['Discounted_price']=df['Cost']*0.1
df
```

```python
#1
df['Discount']=df.apply(lambda row: row.Cost -(row.Cost*0.1),axis=1)
df
```
