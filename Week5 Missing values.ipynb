```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:/Users/mehta/AIML PRACTICES/DataSets/titanic.csv")
```

# a) Check for the missing values

```python
df.isnull().sum()
```

# b) Fill the missing vlues using mean

```python
mean =df['age'].mean()
df['age'] =df['age'].fillna(mean)
df
```

# c) Fill the missing values using median and mode

```python
median = df['age'].median()    
df['age'] = df['age'].fillna(median)
df
```

```python
mode = df['deck'].mode()[0]    
df['deck'] = df['deck'].fillna(mode)
df
```

```python
df['embarked']=df['embarked'].fillna('a')
df
```
