```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential 
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('D:/2. College files/Diploma/5th Sem/CSV FILES/Iris.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
mean=df['SepalWidthCm'].mean()
df['SepalWidthCm']=df['SepalWidthCm'].fillna(mean)
```

```python
df.isnull().sum()
```

```python
x=df.drop(['Species','Id'],axis=1)
y=df['Species']
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(4,),activation='relu',name='input'))
model.add(Dense(10,activation='relu',name='fc2'))
model.add(Dense(3,activation='softmax',name='output'))
```

```python
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
print('NNM Summary:')
print(model.summary())
```

```python
model.fit(x_train,y_train,epochs=13,batch_size=8, verbose=1)
result=model.evaluate(x_test,y_test)
print("accuracy:",result[1])
```
