```python
import pandas as pd
from functools import reduce
data = {
 'Name': ['Alice', 'Bob', 'Charlie', 'David'],
 'Age': [25, 30, 35, 40],
 'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
df
```

```python
##1) map ()
def add(x):
 return x + 2000
Salary_List = df['Salary'].map(add)
print("Added Salaries:\n", Salary_List)
```

```python
##2) filter ()
def get(age):
 if age > 30:
     return True
l1=(df['Age'])    
res=list(filter(get,l1))
print("Grater then 30 years Age :",res)
```

```python
##3)reduce ()
def get(age):
 if age > 30:
     return True
l1=(df['Age'])    
res=list(filter(get,l1))
print("Grater then 30 years Age :",res)
```

```python
# Sum of salary using Lambda
total= reduce(lambda x, y: x + y, df['Salary'])
print("Total Salary:", total)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/titanic.csv")
```

```python
df['age'].plot(kind='box')
plt.show()
```

```python
age_arr = df["age"]
lower_cap = np.percentile(age_arr,1)
upper_cap = np.percentile(age_arr,99)
```

```python
q1=df["age"].quantile(0.25)
q3=df["age"].quantile(0.75)

iqr=q3-q1

upper_bound=q3+(1.5*iqr)
lower_bound=q1-(1.5*iqr)
```

```python
df['age'] = np.where(df['age'] < lower_bound, lower_cap,
              np.where(df['age'] > upper_bound, upper_cap, df['age']))
```

```python
df['age'].plot(kind='box')
plt.show()
```
