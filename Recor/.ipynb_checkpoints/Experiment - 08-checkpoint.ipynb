# Consider the Breast cancer dataset and perform the following operations.

```python
# Import libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# read csv file

df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Breast_Cancer_data.csv")
df.head()
```

```python
# Pre-processing

df.isnull().sum()
```

```python
# Split the data 

x=df.drop(columns=(['id','diagnosis']))
y=df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## a) Build a model using Decision tree algorithm

```python
# Create Decision Tree classifer object
dtc = DecisionTreeClassifier()

# Train Decision Tree Classifer
dtc.fit(X_train,y_train)
```

```python
#Predict the response for test dataset

y_pred = dtc.predict(X_test)
```

## b) Evaluate the model using different metrics

```python
# Model Accuracy, how often is the classifier correct?

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Get predictions
y_pred = dtc.predict(X_test)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Reds)
plt.show()
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True,fmt="d")
```

```python
# Show the Classification_report

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

## c) Compare the Decision tree and Random Forest classifier algorithm

```python
from sklearn.ensemble import RandomForestClassifier  

forest = RandomForestClassifier(n_estimators=100,random_state=0)
forest.fit(X_train,y_train)
```

```python
#Predict the response for test dataset

y_pred = forest.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Cconfusion_matrix 

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
# Classification_report

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
