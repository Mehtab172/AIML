# Create this dataset and perfoem the following

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['Riya', 'Arjun', 'Sneha', 'Rakesh', 'Aditi', 'Kiran', 'Anjali', 'Manish'],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Age': [17, 18, 17, 19, 18, 18, 17, 19],
    'Study_Hours': [3.5, 2.0, 4.0, 1.5, 3.8, 2.5, 4.2, 1.0],
    'Attendance (%)': [95, 82, 98, 70, 92, 85, 99, 65],
    'Marks_Math': [88, 75, 90, 65, 95, 78, 91, 60],
    'Marks_Science': [92, 80, 89, 68, 93, 75, 95, 58],
    'Grade': ['A', 'B', 'A', 'C', 'A', 'B', 'A', 'D']
}
df = pd.DataFrame(data)
df
```

# Univariate Analysis Practice (Matplotlib)
# ðŸ”¹ 1. Plot a histogram of Marks_Math
# ðŸ‘‰ Use plt.hist()

```python
plt.hist(df['Marks_Math'])
plt.title('Histogram')
plt.xlabel('Marks')
plt.ylabel('Number of students')
plt.show()
```

# ðŸ”¹ 2. Draw a bar chart of number of students per Grade
# ðŸ‘‰ Use value_counts() + plt.bar()

```python
grade_counts = df['Grade'].value_counts()
plt.bar(grade_counts.index, grade_counts.values)
plt.title('Number of Students per Grade')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

# ðŸ”¹ 3. Show the distribution of Study_Hours
# ðŸ‘‰ Use plt.boxplot() or plt.hist()


```python
plt.boxplot(df['Study_Hours'])
plt.title('Boxplot of Study Hours')
plt.ylabel('Study Hours')
plt.show()
```

# ðŸ”¹ 4. Plot a pie chart of gender distribution
# ðŸ‘‰ Use plt.pie() with Gender.value_counts()

```python
ff=df['Gender'].value_counts()
plt.pie(ff,labels=ff.index,autopct='%1.1f%%')
plt.show()
```

# ðŸ”¹ 5. Plot the frequency of attendance percentage ranges
# ðŸ‘‰ Categorize attendance: 60â€“70, 70â€“80, 80â€“90, 90â€“100

```python
plt.bar(df['Attendance (%)'].index,df['Attendance (%)'].values)
plt.show()
```

```python
bins = [60, 70, 80, 90, 100]
labels = ['60â€“70', '70â€“80', '80â€“90', '90â€“100']
df['Attendance_Range'] = pd.cut(df['Attendance (%)'], bins=bins, labels=labels, right=True)

attendance_freq = df['Attendance_Range'].value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.bar(attendance_freq.index, attendance_freq.values, color='mediumseagreen')
plt.title('Frequency of Attendance Percentage Ranges')
plt.xlabel('Attendance Range (%)')
plt.ylabel('Number of Students')
plt.show()
```

# Bivariate Analysis Practice (Matplotlib)
# ðŸ”¹ 6. Scatter plot between Study_Hours and Marks_Math
# ðŸ‘‰ Use plt.scatter(x, y)

```python
plt.scatter(df['Study_Hours'],df['Marks_Math'])
plt.show()
```

# ðŸ”¹ 7. Scatter plot between Attendance (%) and Marks_Science

```python
plt.scatter(df['Attendance (%)'],df['Marks_Science'])
plt.title('Study Hours vs Math Marks')
plt.xlabel('Study Hours')
plt.ylabel('Math Marks')
plt.show()
```

# ðŸ”¹ 8. Compare Marks_Math by Gender using side-by-side bar chart
# ðŸ‘‰ Use groupby or filtering and plt.bar()


```python
gender_math = df.groupby('Gender')['Marks_Math'].mean()
plt.bar(gender_math.index,gender_math.values)
plt.title('Average Marks in Math by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Marks in Math')
plt.show()
```

# ðŸ”¹ 9. Line plot showing marks of each student in Math and Science
# ðŸ‘‰ Use plt.plot() with x = names, y = subject marks

```python
plt.plot(df['Name'], df['Marks_Math'], marker='o', label='Math')
plt.plot(df['Name'], df['Marks_Science'], marker='s', label='Science')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.legend()
plt.show()
```

# ðŸ”¹ 10. Boxplot comparing Marks_Math across Grades
# ðŸ‘‰ Use df.boxplot(column='Marks_Math', by='Grade')

```python
df.boxplot(column='Marks_Math', by='Grade')
plt.title('Marks in Math by Grade')
plt.suptitle('')
plt.xlabel('Grade')
plt.ylabel('Marks in Math')
plt.show()
```
