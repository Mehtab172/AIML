```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Mall_Customers.csv")
df.head()
```

```python
df.drop("CustomerID",axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
X
```

```python
std=StandardScaler()
X=std.fit_transform(X)
X
```

```python
import matplotlib.pyplot as plt
wcss=[]
for k in range (1,11):
    km=kmeans=KMeans(n_clusters=k,random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1,11),wcss,marker="o")
```

```python
kmeans=KMeans(n_clusters=5,random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_kmeans,s=50)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="black",marker='X',s=200,label="Centroids")
plt.legend()
```

```python
print("final inertia for k=5",kmeans.inertia_)
```

```python
from sklearn import metrics
print("the selhouette score os this:",metrics.silhouette_score(X,y_kmeans))
```
