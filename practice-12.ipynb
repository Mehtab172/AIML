```python
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt
from tensorflow import keras 
from sklearn import metrics
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train[0].shape
```

```python
plt.matshow(x_train[0])
plt.show()
```

```python
y_train[0]
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train.shape
```

```python
x_train_flattent=x_train.reshape(len(x_train),28*28)
x_test_flattent=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattent.shape
```

```python
model=keras.Sequential([keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_flattent, y_train,epochs=5)
```

```python
model.evaluate(x_test_flattent,y_test)
```

```python
y_predict=model.predict(x_test_flattent)
y_predict[0]
```

```python
plt.matshow(x_test[0])
```

```python
np.argmax(y_predict[0])
```

```python
y_predict_labels=[np.argmax(i) for i in y_predict]
```

```python
y_predict_labels[:5]
```

```python
print("classification report : \n",metrics.classification_report(y_test,y_predict_labels))
```

```python
print("confusion metrics",metrics.confusion_matrix(y_test,y_predict_labels))
```
