```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df['SepalWidthCm'].plot(kind='box')
plt.show()
```

```python
q1=df['SepalWidthCm'].quantile(0.25)
q3=df['SepalWidthCm'].quantile(0.75)
iqr=q3-q1
upper_bound=q1+(1.5*iqr)
lower_bound=q3-(1.5*iqr)

upper_cap=np.percentile(df['SepalWidthCm'],99)
lower_cap=np.percentile(df['SepalWidthCm'],1)
```

```python
m=df['SepalWidthCm'].mean()
print(m)
```

```python
for i in df['SepalWidthCm']:
    if i<lower_bound or i>upper_bound:
        df['SepalWidthCm']=df['SepalWidthCm'].replace(i,m)
        
```
