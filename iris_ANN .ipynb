```python
import numpy as np 
import pandas as pd   
from sklearn.preprocessing import OneHotEncoder  
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Iris.csv")
df.head()
```

```python
df = df.drop("Id", axis=1)
```

```python
x=df.drop("Species",axis=1) 
y=df["Species"] 
```

```python
# One-hot encode the target column
y= pd.get_dummies(y)
y
```

```python
# Step 5: Split train/test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 6: Build ANN model
model = Sequential()  
model.add(Dense(10, input_shape=(4,), activation='relu', name="input"))   # 4 features
model.add(Dense(10, activation='relu',name="fc2"))                         
model.add(Dense(3, activation='softmax',name="output"))                     
```

```python
# Step 7: Compile modelabs
optimizer=Adam(learning_rate=0.0001)
model.compile(optimizer,loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
print("Neural Network Model Summary")
print(model.summary())
```

```python
# Train
model.fit(x_train,y_train,epochs=200)
```

```python
result=model.evaluate(x_test,y_test)
print("Final test set Accuracy",result[1])
```

```python
# Example custom input (SepalLength, SepalWidth, PetalLength, PetalWidth)
x_test1 = np.array([[4.6, 3.4, 1.4, 0.3]])

# Predict probabilities for each class
y_pred = model.predict(x_test1)
```

```python
# Fit label encoder on original target column
label_encoder = LabelEncoder()
label_encoder.fit(df["Species"])   # 👈 Now it remembers mapping
```

```python
# Get class index (0,1,2)
pred_class = np.argmax(y_pred, axis=1)

# Map back to species names
print("Predicted probabilities:", y_pred)
print("Predicted class index:", pred_class)
print("Predicted species:", label_encoder.inverse_transform(pred_class))
```
