```python
import re
import string
```

```python
stop_words=["the","is","and","of","to"]
```

```python
punctuation_regex = re.compile(r"[,.!?]")
```

```python
def drop_stop_words_and_punctuations(text):
    words = text.split()
    filtered_words = []
    for word in words:
        if word not in stop_words and not punctuation_regex.match(word):
          filtered_words.append(word)
    return " ".join (filtered_words)
```

```python
text="My Self Mehtab Ahmed I am studying in 5th semester cse , the cse department is very good ! ."
filtered_text = drop_stop_words_and_punctuations(text)
```
