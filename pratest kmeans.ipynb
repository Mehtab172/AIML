```python
import numpy as np 
import pandas as pd   
import matplotlib.pyplot as plt   
from sklearn.preprocessing import StandardScaler   
from sklearn.cluster import KMeans  
```

```python
df=pd.read_csv("C:/Users/mehta/AIML-20CS51I/DataSets/Mall_Customers.csv")
```

```python
df.head()
```

```python
x=df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
x.head()
```

```python
std=StandardScaler()
x=std.fit_transform(x)
```

```python
x[:5]
```

```python
wcss=[]
for k in  range(1,11):
    km=KMeans(n_clusters=k,random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker='*')
plt.title("Elbow Method")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=y_kmeans)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='X',label="Centroids",c='black',s=200)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("K-Means Clustering (Mall Customers)")
```

```python
print("inertia for k5 is ",kmeans.inertia_)
```

```python
from sklearn import metrics
```

```python
print("siloutte for this k5 is",metrics.silhouette_score(x,y_kmeans))
```
