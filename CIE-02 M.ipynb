```python
q1=df['fare'].quantile(0.25)
q3=df['fare'].quantile(0.75)
iqr=q3-q1
upper_bound=q3+(1.5*iqr)
lower_bound=q1-(1.5*iqr)
```

```python
m=df['fare'].mean()
for i in df['fare']:
    if i<lower_bound or i>upper_bound:
        df['fare']=df['fare'].replace(i,m)
```
